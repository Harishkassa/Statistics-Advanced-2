5. What is the difference between a Z-test and a T-test? Explain when to use
each.


-> Population Variance Known?

Z-test: Used when population variance (σ²) is known.

T-test: Used when population variance is unknown.

Sample Size:

Z-test: Preferred for large samples (n ≥ 30).

T-test: Preferred for small samples (n < 30).

Distribution Used:

Z-test: Based on Standard Normal Distribution (Z-distribution).

T-test: Based on Student’s t-distribution.



6. Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10
p = 0.5

# Generate binomial distribution samples
data = np.random.binomial(n, p, size = 100000)

# Plot histogram
plt.hist(data, bins=11)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.show()
```

9. Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)

    # Z-score formula
    z_scores = [(x - mean) / std for x in data]
    return z_scores

# Example dataset (random normal data)
np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=1000)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
plt.figure(figsize=(8, 5))
plt.hist(z_scores, bins=30, color='skyblue')
plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
plt.axvline(1, color='green', linestyle='--', label='1 Std Dev Above (Z=1)')
plt.axvline(-1, color='green', linestyle='--', label='1 Std Dev Below (Z=-1)')

plt.title("Histogram of Standardized Data (Z-scores)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
